```python
Fibonnacci sequence is generated by adding previous last two numbers .by starting with 1 and 2,the first terms will be 1,2,3,5,8,13,21,34,55..... 
```

```python
n=int(input("enter value of n="))
x=0
y=1 
z=x+y
while z<n:
    print(z,end=' ')
    x=y
    y=z 
    z=x+y
```
